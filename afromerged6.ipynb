```python
from rpy2.robjects import r, pandas2ri
filename = "merged_r6_data_2016_36countries2.sav"
w = r('foreign::read.spss("%s", to.data.frame=TRUE,use.value.labels = FALSE)' % filename)
cols = list(r("attr")(w, "variable.labels")) 
w.colnames = cols
w.to_csvfile("merged_r6_data_2016_36countries.csv")
```
